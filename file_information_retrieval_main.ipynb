### Main method

```python
def save_paths(base_dir, file_end, save_file_name="wav_paths.csv"):
  wav_files = []
  #Iterating over all file-paths 
  for root, dirs, files in os.walk(base_dir):
      for file in files:
          if file.endswith(str(file_end)):
            #Storing each parth of the path string to a variable
            path = os.path.join(root, file)
            leak_present = 1 if "_niO_" in file else 0
            file_split = file.split("_")
            recording = file_split[0]
            knob_rotations = file_split[2]
            mic = file_split[3]
            root_split = root.split(base_dir)[1].split("/")
            leak_type = root_split[0]
            environment = root_split[1]
            environment_folder = root_split[2]
            #Saving all the variables as an array in the wav_files array 
            wav_files.append([path,leak_type,environment,recording,mic,knob_rotations,leak_present])
          

  #Saving the file
  if ".csv" not in save_file_name:
    save_file_name += ".csv"

  with open(save_file_name,'w') as result_file:
      wr = csv.writer(result_file, dialect='excel')
      wr.writerow(["path","leak_type","environment","recording","mic","knob_rotations","leak_present"])
      for wav_file in wav_files:
        wr.writerow(wav_file)
```

```python
#Calling the method
save_paths("/content/gdrive/MyDrive/Masteroppgave/IDMT_dataset/", ".wav", "wav_paths_master_TEST.csv")
```
